# Finite State Machines

States: A, B, C, D

Starting state: A, Terminating state: D

States and state transitions.

image.png

```python
current_state = 'A'
print('A is the starting state.')
while current_state!='D':
    print('Current State is',current_state)
    n = int(input('Enter transition input(0,1): '))
    if current_state=='A':
        if n==1:
            current_state = 'B'
        if n==0:
            current_state = 'C'
        continue
    if current_state=='B':
        if n==1:
            current_state = 'A'
        if n==0:
            current_state = 'D'
        continue
    if current_state=='C':
        if n==1:
            current_state = 'D'
        if n==0:
            current_state = 'A'
        continue
print('Reach state D, which is the terminating state.')
```
